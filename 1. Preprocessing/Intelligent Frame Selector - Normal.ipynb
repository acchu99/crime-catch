```python
import os
import cv2
```

```python
from deepstack_sdk import ServerConfig, Detection
config = ServerConfig("http://localhost:80")
detection = Detection(config)
```

```python
def saveFrames(name):
    person = False
    cap = cv2.VideoCapture(f'./Normal/{name}.mp4')
    i, n = 1, 30
    j = 1
    if cap.isOpened()==False:
        print("Either file not found or wrong codec used")
    while cap.isOpened():
        ret, frame = cap.read()
        if ret == False:
            break
        if i%n==0:
            i+=1
            response = detection.detectObject(frame,output=None)
            for obj in response:
                if obj.label=="person" and obj.confidence>0.70:
                    cv2.imwrite(f'./Frames/Normal/{name}_fr{str(j)}.jpg',frame)
                    j+=1
                else:
                    pass
        else:
            i+=1
    cap.release()
    cv2.destroyAllWindows()
```

```python
normal = [normal.split(".")[0] for normal in os.listdir("./Normal")]
print(len(normal))
```

```python
length = len(normal)
i=1

for norm in normal:
    saveFrames(norm)
    print(f"{i} file done.")
    i+=1
```
